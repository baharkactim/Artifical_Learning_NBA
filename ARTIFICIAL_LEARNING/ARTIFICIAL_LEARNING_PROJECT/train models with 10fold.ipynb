```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import KFold
from sklearn import metrics
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
data = pd.read_csv("chiago-stats-preprocessed.csv")
```

```python
data
```

```python
X=data.iloc[:,:-1].values
y=data['changes_of_win_rating'].values
```

```python
#separates data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Decision Tree Algorithm
decision_model = DecisionTreeClassifier()

#KNN Algorithm
knn_model = KNeighborsClassifier()

#Logistic Regression Algorithm
logistic_model = LogisticRegression()

#Naive Bayes Algorithm
naive_model = GaussianNB()

#Random Forest Algorithm
random_model = RandomForestClassifier()

#AdaBoost Algorithm 
ada_model = AdaBoostClassifier()

#SV Classifier Algorithm
sv_model = SVC()

#Multi-Layer Perceptron Algorithm
multi_model = MLPClassifier()
```

```python
def scoreResults(model, X_train, X_test, y_train, y_test):

    y_train_predict = model.predict(X_train)
    y_test_predict = model.predict(X_test)

    r2_train = metrics.r2_score(y_train, y_train_predict)
    r2_test = metrics.r2_score(y_test, y_test_predict)

    mse_train = metrics.mean_squared_error(y_train, y_train_predict)
    mse_test = metrics.mean_squared_error(y_test, y_test_predict)
    
    acc = accuracy_score(y_test, y_test_predict)

    return [r2_train, r2_test, mse_train, mse_test, y_train_predict, y_test_predict, acc]
```

```python
def dic2bar(title, dic):
    keys = dic.keys()
    values = dic.values()
    plt.xticks(rotation=90)
    plt.title(title)
    plt.bar(keys, values)
    plt.show()
```

```python
def regressionAlgorithm(reg_model, X_train, X_test, y_train, y_test, reg_name):
    model = reg_model
    k = 10
    iter = 1
    cv = KFold(n_splits=k, random_state = 0, shuffle=True)
    print(reg_name, "Scores")
    
    acc_values = {}
    train_mse_values = {}
    test_mse_values = {}
    train_r2_values = {}
    test_r2_values = {}
    
    for train_index, test_index in cv.split(X):
        X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
        model.fit(X_train, y_train)
        
        result = scoreResults(model = model
                          ,X_train = X_train
                          ,X_test = X_test
                          ,y_train = y_train
                          ,y_test = y_test)
        
        print(f"{iter}. veri kesiti")
        print(f"Train R2 Score: {result[0]:.4f} MSE: {result[2]:.4f}")
        print(f"Test R2 Score: {result[1]:9.4f} MSE: {result[3]:.4f}\n")
        print("Model Accuracy = ", result[-1])
        
        acc_values[str(iter) + ". veri kesiti"] = result[-1]
        train_mse_values[str(iter) + ". veri kesiti"] = result[2]
        test_mse_values[str(iter) + ". veri kesiti"] = result[3]
        train_r2_values[str(iter) + ". veri kesiti"] = result[0]
        test_r2_values[str(iter) + ". veri kesiti"] = result[1]
        
        iter += 1
        
        #Residual Plot
        plt.scatter(np.arange(len(y_test)), y_test, color = "red", label = "Exact Values")
        plt.scatter(np.arange(len(result[-2])), result[-2], color = "blue", label = "Predicted Value")
        #plt.hlines(y=0, xmin=0, xmax=80, color= 'blue')
        plt.title(reg_name)
        plt.ylabel('Values')
        plt.xlabel('Indices')
        plt.legend()
        plt.show()
    
    dic2bar(reg_name + " Accuracy Scores", acc_values)
    print(acc_values)
    dic2bar(reg_name + " Train MSE Values", train_mse_values)
    print(train_mse_values)
    dic2bar(reg_name + " Test MSE Values", test_mse_values)
    print(test_mse_values)
    dic2bar(reg_name + " Train R2 Values", train_r2_values)
    print(train_r2_values)
```

```python
regressionAlgorithm(decision_model, X_train, X_test, y_train, y_test, 'Decision Tree Algorithm')
```

```python
regressionAlgorithm(knn_model, X_train, X_test, y_train, y_test, 'KNN Algorithm')
```

```python
regressionAlgorithm(logistic_model, X_train, X_test, y_train, y_test, 'Logistic Regression Algorithm')
```

```python
regressionAlgorithm(naive_model, X_train, X_test, y_train, y_test, 'Naive Bayes Algorithm')
```

```python
regressionAlgorithm(random_model, X_train, X_test, y_train, y_test, 'Random Forest Algorithm')
```

```python
regressionAlgorithm(ada_model, X_train, X_test, y_train, y_test, 'AdaBoost Algorithm ')
```

```python
regressionAlgorithm(sv_model, X_train, X_test, y_train, y_test, 'SV Classifier Algorithm')
```

```python
regressionAlgorithm(multi_model, X_train, X_test, y_train, y_test, 'Multi-Layer Perceptron Algorithm')
```
