# combine csvs for features

```python
features = ["block-pct", "defensive-efficiency", "offensive-efficiency", "steal-pct", "total-rebounds-per-game", "true-shooting-percentage", "turnover-pct", "win-pct-all-games"]
```

```python
for feature in features:
    os.chdir(root + "\\nba-stat-data\\" + feature)
    file_extension = '.csv'
    all_filenames = [i for i in glob.glob(f"*{file_extension}")]
    data = pd.concat([pd.read_csv(f) for f in all_filenames])
    
    data = data.rename(columns={"2020": feature + "-year", "last3": feature + "-last-3", "last1": feature + "-last-1", "home": feature + "-home", "away": feature + "-away", "2019": feature + "-last-year"})
    data.to_csv(root+ "\\nba-stat-data\\" + feature + ".csv", index=False)
```

# preprocessing

```python
data = pd.read_csv("chiago-stats-data.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
for columnName in data.columns:
    for rowIndex in range(data.shape[0]):
        if(data[columnName].loc[rowIndex] == "--"):
            data[columnName].loc[rowIndex] = None
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
data = data.reset_index(drop=True)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
columns = list(data.columns)
columns
```

```python
columns.pop(0)
columns
```

```python
for columnName in columns:
    for rowIndex in range(data.shape[0]):
        #print(str(data[columnName][rowIndex]))
        if str(data[columnName][rowIndex]).find("%") != -1:
            data[columnName][rowIndex] = float(data[columnName][rowIndex].replace("%", ""))
        else:
            data[columnName][rowIndex] = float(data[columnName][rowIndex])
```

```python
data
```

```python
data[["year", "month", "day"]] = data["date"].str.split("-", expand = True)
del data["date"]
```

```python
data.to_csv("chiago-stats-data-preprocessed.csv", index=False)
```

# playing with features

```python
data = pd.read_csv("chiago-stats-data-preprocessed.csv")
```

```python
data.head()
```

```python
mac_sonuclari = data["win-pct-all-games-last-1"].values
```

```python
win_rating = []
for mac_sayisi in range(1, len(mac_sonuclari)+1):
    su_ana_kadar_olan_maclar = mac_sonuclari[0:mac_sayisi]
    ones = 0
    for onceki_maclar in su_ana_kadar_olan_maclar:
        if onceki_maclar == 1:
            ones += 1
    print(str(mac_sayisi) + ". maç: " + str(mac_sonuclari[mac_sayisi-1]) + " kazanma oranım: " + str(ones/mac_sayisi))
    win_rating.append(float(ones/mac_sayisi))
```

```python
data["win_rating"] = win_rating
```

```python
data
```

```python
data = data.drop(["win-pct-all-games-year", "win-pct-all-games-last-3", "win-pct-all-games-last-1", "win-pct-all-games-home", "win-pct-all-games-away", "win-pct-all-games-last-year"], axis=1)
```

```python
data.to_csv("chiago-stats-win-rating.csv", index=False)
```

```python
win_rating = data["win_rating"].values
win_rating
```

classes:
    1: win oranı arttı
    0: win oranı aynı
    -1: win oranı düştü

```python
changes_of_win_rating = [0]
for i in range(1, len(win_rating)):
    if win_rating[i] > win_rating[i-1]:
        changes_of_win_rating.append(1)
    elif win_rating[i] < win_rating[i-1]:
        changes_of_win_rating.append(-1)
    elif win_rating[i] == win_rating[i-1]:
        changes_of_win_rating.append(0)
```

```python
data["changes_of_win_rating"] = changes_of_win_rating
del data["win_rating"]
```

```python
data.to_csv("chiago-stats-changes-of-win-rating.csv", index=False)
```
