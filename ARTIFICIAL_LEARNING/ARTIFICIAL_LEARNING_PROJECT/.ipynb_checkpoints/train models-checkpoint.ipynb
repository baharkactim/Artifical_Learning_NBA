```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
data = pd.read_csv("chiago-stats-preprocessed.csv")
```

```python
data
```

```python
X=data.iloc[:,:-1].values
y=data['changes_of_win_rating'].values
```

```python
#separates data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Decision Tree Algorithm
decision_model = DecisionTreeClassifier()

#KNN Algorithm
knn_model = KNeighborsClassifier()

#Logistic Regression Algorithm
logistic_model = LogisticRegression()

#Naive Bayes Algorithm
naive_model = GaussianNB()

#Random Forest Algorithm
random_model = RandomForestClassifier()

#AdaBoost Algorithm 
ada_model = AdaBoostClassifier()

#SV Classifier Algorithm
sv_model = SVC()

#Multi-Layer Perceptron Algorithm
multi_model = MLPClassifier()
```

```python
def regressionAlgorithm(reg_model, X_train, X_test, y_train, y_test, reg_name):
    model = reg_model
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    #Residual Plot
    plt.scatter(np.arange(len(y_test)), y_test, color = "red", label = "Exact Values")
    plt.scatter(np.arange(len(y_test_pred)), y_test_pred, color = "blue", label = "Predicted Value")
    #plt.hlines(y=0, xmin=0, xmax=80, color= 'blue')
    plt.title(reg_name)
    plt.ylabel('Values')
    plt.xlabel('Indices')
    plt.legend()
    plt.show()
    
    print(reg_name, "Scores")
    print("Train MSE =", mean_squared_error(y_train, y_train_pred))
    print("Train R2 =", r2_score(y_train, y_train_pred))
    print("Test MSE =", mean_squared_error(y_test, y_test_pred))
    print("Test R2 =", r2_score(y_test, y_test_pred))
    print("Model Accuracy = ", accuracy_score(y_test, y_test_pred))
    
    acc_values[reg_name] = accuracy_score(y_test, y_test_pred)
    train_mse_values[reg_name] = mean_squared_error(y_train, y_train_pred)
    test_mse_values[reg_name] = mean_squared_error(y_test, y_test_pred)
    train_r2_values[reg_name] = r2_score(y_train, y_train_pred)
    test_r2_values[reg_name] = r2_score(y_test, y_test_pred)
    
    print("------------------------------------------------------------------")
```

```python
def dic2bar(title, dic):
    keys = dic.keys()
    values = dic.values()
    plt.xticks(rotation=90)
    plt.title(title)
    plt.bar(keys, values)
```

```python
acc_values = {}
train_mse_values = {}
test_mse_values = {}
train_r2_values = {}
test_r2_values = {}
```

```python
regressionAlgorithm(decision_model, X_train, X_test, y_train, y_test, 'Decision Tree Algorithm')
```

```python
regressionAlgorithm(knn_model, X_train, X_test, y_train, y_test, 'KNN Algorithm')
```

```python
regressionAlgorithm(logistic_model, X_train, X_test, y_train, y_test, 'Logistic Regression Algorithm')
```

```python
regressionAlgorithm(naive_model, X_train, X_test, y_train, y_test, 'Naive Bayes Algorithm')
```

```python
regressionAlgorithm(random_model, X_train, X_test, y_train, y_test, 'Random Forest Algorithm')
```

```python
regressionAlgorithm(ada_model, X_train, X_test, y_train, y_test, 'AdaBoost Algorithm ')
```

```python
regressionAlgorithm(sv_model, X_train, X_test, y_train, y_test, 'SV Classifier Algorithm')
```

```python
regressionAlgorithm(multi_model, X_train, X_test, y_train, y_test, 'Multi-Layer Perceptron Algorithm')
```

```python
dic2bar("Accuracy Scores", acc_values)
```

```python
dic2bar("Train MSE Values", train_mse_values)
```

```python
dic2bar("Test MSE Values", test_mse_values)
```

```python
dic2bar("Train R2 Values", train_r2_values)
```

```python
dic2bar("Test R2 Values", test_r2_values)
```

# predictions

```python
def makePrediction(lastGameCount=data.shape[0], year=2021, month=6, day=2, model=random_model):
    featureCount = (len(data.iloc[0])-4)

    prediction = [0]*featureCount # 4: year, month, day

    for i in range(-lastGameCount, -1):
        currentData = data.iloc[i]
        currentData = currentData.drop(["year", "month", "day", "changes_of_win_rating"])
        for index in range(len(currentData)):
            prediction[index] += currentData[index]
            
    prediction = [feature/lastGameCount for feature in prediction]
    prediction.append(year)
    prediction.append(month)
    prediction.append(day)
    
    pred = model.predict(pd.Series(prediction).values.reshape(1, -1))[0]
    return str("date: " + str(year) + "-" + str(month) + "-" + str(day) + ", last " + str(lastGameCount) + " game"), pred
```
