```python
from bs4 import BeautifulSoup
import requests
from datetime import date as dt, timedelta
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time
import re
```

```python
path = r"C:\Users\Soner\Downloads\chromedriver_win32\chromedriver.exe"
options = Options()
options.headless = True
```

```python
teams = ["Chicago"]
column_names = ["2020", "last3", "last1", "home", "away", "2019"]
```

```python
base_url = "https://www.teamrankings.com"
r = requests.get("https://www.teamrankings.com/nba/stats/")
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
team_stats = soup.find_all("ul", attrs={"class": "chooser-list"})[0]
```

```python
features_links = [a["href"] for a in team_stats.find_all("a") if a["href"] != "#"]
```

```python
for i in range(len(features_links)):
    print(str(i) + "- " + features_links[i])
```

```python
feature_index = 2
```

```python
sdate = dt(2013, 5, 24)   # start date2013-05-23
edate = dt(2015, 12, 31)   # end date

delta = edate - sdate
dicts = []
error_days = []
error_links = []

driver = webdriver.Chrome(executable_path=path, options=options)
for i in range(delta.days + 1):
    date = str(sdate + timedelta(days=i))
    driver.get(base_url + features_links[feature_index] + "?date=" + date)
    time.sleep(1)
    rows = driver.find_element_by_class_name("tr-table.datatable.scrollable.dataTable.no-footer").find_elements_by_tag_name("tr")
    rows_indices = []
    for index in range(len(rows)):
        row_content = rows[index].get_attribute("innerHTML")
        soup_row = BeautifulSoup(row_content, 'html.parser')
        try:
            team_name = soup_row.find("a").text
            for team in teams:
                if team == team_name:
                    rows_indices.append(index)
        except:
            pass
    for index in rows_indices:
        row_content = rows[index].get_attribute("innerHTML")
        soup_row = BeautifulSoup(row_content, 'html.parser')
        team_name = soup_row.find("a").text
        lst = str(soup_row).split("</td>")
        data = {
            "date": date,
            column_names[0]: lst[2].split(">")[1],
            column_names[1]: lst[3].split(">")[1],
            column_names[2]: lst[4].split(">")[1],
            column_names[3]: lst[5].split(">")[1],
            column_names[4]: lst[6].split(">")[1],
            column_names[5]: lst[7].split(">")[1],
        }
        dicts.append(data)
        print(data)
driver.quit()

```

```python
import csv
csv_file = "C:/Users/soner/Desktop/works/" + features_links[feature_index].replace("/", "-") + "-" + str(sdate) + "_to_" + str(edate) + ".csv"
csv_columns = ["date", "2020", "last3", "last1", "home", "away", "2019"]
try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in dicts:
            writer.writerow(data)
except IOError:
    print("I/O error")
```
